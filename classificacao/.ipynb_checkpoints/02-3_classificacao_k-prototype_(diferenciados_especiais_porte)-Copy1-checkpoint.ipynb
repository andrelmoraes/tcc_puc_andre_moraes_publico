### 4.2 Normalizando a escala


---
Conforme visto no notebook `01_tratamento_de_dados_classificacao-final`, o dataset possui diversos outliers e uma grande dispersão dos dados, entretanto, como os outliers são representativos do setor econômico e o acompanhamento de sua arrecadação é essencial para a atividade da RFB, serão mantidos. Junto a isso, alguns campos, como `pj_receita_bruta` possuem uma escala muito maior que outros campos, de modo que é possível que o primeiro distorça a classificação.

Diante disso, normalizarei o dataset antes da classificação.

A normalização incluirá os campos categóricos convertidos em binários no dataset. Testes realizados anteriormente mostraram não haver diferença em normalizar o dataset com todos os valores ou normalizar apenas os valores numéricos e depois concatenar os valores categóricos.

---

```python
from sklearn.preprocessing import Normalizer

dc_normalizer = Normalizer()

dataset_cluster_normalized = dc_normalizer.fit_transform(dataset_cluster)

print(dataset_cluster_normalized)
```

### 4.3 Verificando o número ótimo de clusters com o Método do Cotovelo (Elbow Method)

---

Para determinar o número ótimo de clusters a partir do dataset normalizado, vou usar o Método do Cotovelo, buscando o ponto onde há inflexão da curva, testando de 2 a 15 clusters.

O algoritmo K-Prototypes implementado em Python possui as mesmas interfaces do K-Means, de modo que o código gerado será similar, com a diferença de que o K-Prototypes gera o custo de cada agrupamento ao invés da inércia.

O algoritmo K-Prototypes classifica os elementos do dataset a partir da definição de protótipos que agregam as características quantitativas (através de uma medida de distância) e as qualitativas (através de uma medida de similaridade) e definem um centróide. Após a seleção inicial dos centróides, todos os elementos do dataset são alocados a um cluster e a medida de similaridade é testada novamente contra os protótipos e, se algum elemento estiver mais próximo de outro protótipo, os clusters são atualizados e os protótipos recalculados. O algoritmo transcorre até que não haja mais realocações a serem feitas.

Na saída do processamento abaixo, a iteração marcada com a melhor é a última que houve realocação.

Em relação ao hiperparâmetro `init`, foram testados os dois disponíveis na implementação em Python (`'Huang` e `'Cao'`) e a inicialização `'Cao'` apresentou melhores resultados, sem clusters com poucos ou nenhum elemento. 


---

```python
from kmodes.kprototypes import KPrototypes
cost = []


for i in range(2,15):
    kprototypes = KPrototypes(n_clusters = i, init = 'Cao',verbose = 1, n_init = 6, n_jobs = 12, random_state = 406)
    kprototypes.fit(dataset_cluster_normalized, categorical=[0,1,15,16,17])
    cost.append(kprototypes.cost_)
    
```

```python
plt.plot(range(2, 15), cost)
plt.grid()
plt.title('Método do Cotovelo')
plt.xlabel('Número de Clusters')
plt.ylabel('Custo total')
plt.savefig('02-3_k-proto_metodo_cotovelo.png')
plt.show()
```

---
Considerando que o teste com o Método do Cotovelo não foi conclusivo, já que é possível que o número se situe entre 3 e 5 a 8 clusters, e que o método de Análise de Silhueta, para tentar identificar o número ótimo de clusters, não se aplica à classificação em tela porque o algoritmo K-Prototypes usa duas medidas distintas para agregação (uma para os valores quantitativos e outra para os valores categóricos), irei analisar os resultados para 3, 4, 5, 6, 7 e 8 clusters.

---
Criando um DataFrame que conterá os dados agregados de cada cluster para a realização de análises das características de cada cluster gerado.

___

```python
resumo_clusters_3 = pd.DataFrame(columns = ['cluster'])

resumo_clusters_3.cluster = ['cluster 0', 'cluster 1', 'cluster 2']

# Totalizadores dos clusters

receita_bruta_total = sum(dataset_final_kproto3.pj_receita_bruta)
arrecadacao_total = sum(dataset_final_kproto3.pj_arrecadacao)
saldo_pagar_total = sum(dataset_final_kproto3.dctf_valor_saldo_pagar)
credito_perdcomp_total = sum(dataset_final_kproto3.credito_total_perdcomp)
devedor_total = sum(dataset_final_kproto3.debitos_total_devedor)
judicial_total = sum(dataset_final_kproto3.debitos_total_susp_judicial)
admin_total = sum(dataset_final_kproto3.debitos_total_susp_admin)
debitos_total = (sum(dataset_final_kproto3['debitos_total_susp_judicial']) 
                 + sum(dataset_final_kproto3['debitos_total_susp_admin']) 
                 + sum(dataset_final_kproto3['debitos_total_devedor'] ))

for i in range(0,3):
    
    # Contagem de contribuintes por clusters
    
    resumo_clusters_3.at[i, 'contribuintes'] = len(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i])
    
    # Dados agregados de Receita do Bruta
    
    pj_receita_bruta = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].pj_receita_bruta)
    
    resumo_clusters_3.at[i,'receita_bruta_total'] = pj_receita_bruta
    
    # Dados agregados de Arrecadação
    
    pj_arrecadacao = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].pj_arrecadacao)
    
    resumo_clusters_3.at[i,'arrecadacao_total'] = pj_arrecadacao
    
    
    
    # Dados agregados de Saldo a Pagar
    
    saldo_pagar = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].dctf_valor_saldo_pagar)
    
    resumo_clusters_3.at[i,'saldo_pagar_total'] = saldo_pagar
    
    
    
    # Dados agregados de créditos apresentados em PER/DCOMP
    
    credito_perdcomp = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].credito_total_perdcomp)
    
    resumo_clusters_3.at[i,'dcomp_total'] = credito_perdcomp
    

    # Dados agregados de débitos exigíveis em 31/12/2019
    
    devedor = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].debitos_total_devedor)
    
    resumo_clusters_3.at[i,'devedor_total'] = devedor
    
    
    # Dados agregados de débitos suspensos judicialmente em 31/12/2019
    
    judicial = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].debitos_total_susp_judicial)
    
    resumo_clusters_3.at[i,'judicial_total'] = judicial
    
    
    # Dados agregados de débitos suspensos administrativamente em 31/12/2019
    
    admin = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].debitos_total_susp_admin)
    
    resumo_clusters_3.at[i,'admin_total'] = admin
    
    
    # Dados agregados de débitos totais constituídos em 31/12/2019
    
    total = (sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].debitos_total_susp_admin) +
             sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].debitos_total_susp_judicial) +
             sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].debitos_total_devedor))    
    
    resumo_clusters_3.at[i,'debitos_total'] = total
 
    resumo_clusters_3.at[i,'Grande'] = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].Porte_Grande)
    resumo_clusters_3.at[i,'Médio'] = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].Porte_Médio)
    resumo_clusters_3.at[i,'Pequeno'] = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].Porte_Pequeno)
    resumo_clusters_3.at[i,'Diferenciados'] = (
        sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].status_diferenciada) -
        sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].status_especial))
    resumo_clusters_3.at[i,'Especiais'] = sum(dataset_final_kproto3[dataset_final_kproto3['cluster_contrib']==i].status_especial)
    
   
resumo_clusters_3.head()
```

```python
resumo_clusters_8 = pd.DataFrame(columns = ['cluster'])

resumo_clusters_8.cluster = ['cluster 0', 'cluster 1', 'cluster 2', 'cluster 3', 'cluster 4', 'cluster 5', 'cluster 6', 'cluster 7']

# Totalizadores dos clusters

receita_bruta_total = sum(dataset_final_kproto8.pj_receita_bruta)
arrecadacao_total = sum(dataset_final_kproto8.pj_arrecadacao)
saldo_pagar_total = sum(dataset_final_kproto8.dctf_valor_saldo_pagar)
credito_perdcomp_total = sum(dataset_final_kproto8.credito_total_perdcomp)
devedor_total = sum(dataset_final_kproto8.debitos_total_devedor)
judicial_total = sum(dataset_final_kproto8.debitos_total_susp_judicial)
admin_total = sum(dataset_final_kproto8.debitos_total_susp_admin)
debitos_total = (sum(dataset_final_kproto8['debitos_total_susp_judicial']) 
                 + sum(dataset_final_kproto8['debitos_total_susp_admin']) 
                 + sum(dataset_final_kproto8['debitos_total_devedor'] ))

for i in range(0,8):
    
    # Contagem de contribuintes por clusters
    
    resumo_clusters_8.at[i, 'contribuintes'] = len(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i])
    
    # Dados agregados de Receita do Bruta
    
    pj_receita_bruta = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].pj_receita_bruta)
    
    resumo_clusters_8.at[i,'receita_bruta_total'] = pj_receita_bruta
    
    # Dados agregados de Arrecadação
    
    pj_arrecadacao = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].pj_arrecadacao)
    
    resumo_clusters_8.at[i,'arrecadacao_total'] = pj_arrecadacao
       
    # Dados agregados do valor declarado de débitos
    
    valor_debito = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].dctf_valor_debito)
    resumo_clusters_8.at[i,'valor_debito_total'] = valor_debito
    
    # Dados agregados de valores compensados
    
    resumo_clusters_8.at[i,'valor_compensado'] = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].dctf_valor_comp_dcomp)
    
       
    # Dados agregados de Saldo a Pagar
    
    saldo_pagar = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].dctf_valor_saldo_pagar)
    
    resumo_clusters_8.at[i,'saldo_pagar_total'] = saldo_pagar
    
    
    # Dados agregados de créditos apresentados em PER/DCOMP
    
    credito_perdcomp = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].credito_total_perdcomp)
    
    resumo_clusters_8.at[i,'dcomp_total'] = credito_perdcomp
    
    # Dados agregados de débitos exigíveis em 81/12/2019
    
    devedor = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].debitos_total_devedor)
    
    resumo_clusters_8.at[i,'devedor_total'] = devedor
    
    # Dados agregados de débitos suspensos judicialmente em 31/12/2019
    
    judicial = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].debitos_total_susp_judicial)
    
    resumo_clusters_8.at[i,'judicial_total'] = judicial
    
    
    # Dados agregados de débitos suspensos administrativamente em 31/12/2019
    
    admin = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].debitos_total_susp_admin)
    
    resumo_clusters_8.at[i,'admin_total'] = admin
   
    
    # Dados agregados de débitos totais constituídos em 31/12/2019
    
    total = (sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].debitos_total_susp_admin) +
             sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].debitos_total_susp_judicial) +
             sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].debitos_total_devedor))    
    
    resumo_clusters_8.at[i,'debitos_total'] = total
    
    
    resumo_clusters_8.at[i,'Grande'] = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].Porte_Grande)
    resumo_clusters_8.at[i,'Médio'] = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].Porte_Médio)
    resumo_clusters_8.at[i,'Pequeno'] = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].Porte_Pequeno)
    resumo_clusters_8.at[i,'Diferenciados'] = (
        sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].status_diferenciada) -
        sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].status_especial))
    resumo_clusters_8.at[i,'Especiais'] = sum(dataset_final_kproto8[dataset_final_kproto8['cluster_contrib']==i].status_especial)
    


resumo_clusters_8.head(10)
```
