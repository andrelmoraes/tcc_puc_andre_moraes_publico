# Modelo Auto-Arima

---
O modelo implementado a seguir servirá de base para a avaliação das predições dos demais modelos. Utilizarei apenas o campo arrecadação para fazer a predição.

___

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.dates as dts
import pandas as pd
import math
import requests
import json
import os
from datetime import datetime
import math
from sklearn.metrics import mean_squared_error

import warnings
warnings.filterwarnings("ignore")

from statsmodels.tools.sm_exceptions import ConvergenceWarning
warnings.simplefilter('ignore', ConvergenceWarning)
```

## 1. Importando os datasets

```python
dataset_cluster2_series = pd.read_csv("datasets//dataset_cluster2_series.csv")
dataset_cluster2_validation = pd.read_csv("datasets//dataset_cluster2_validation.csv")
```

```python
dataset_cluster2_series.head(20)
```

```python
dataset_cluster2_validation.head(20)
```

```python
## Transformando o campo mes em índice


serie_cluster2 = dataset_cluster2_series.drop(['mes'], axis=1)
serie_cluster2.index = dataset_cluster2_series.mes

validacao_cluster2 = dataset_cluster2_validation.drop(['mes'], axis=1)
validacao_cluster2.index = dataset_cluster2_validation.mes

# Excluindo os campos que não serão usados no modelo AutoArima

serie_cluster2 = serie_cluster2['arrecadacao']
validacao_cluster2 = validacao_cluster2['arrecadacao']
```

## 2. Implementando o modelo Auto-Arima

```python
#Função para calculo do MAPE - Mean Absolute Percentage Error, que será  utilizado como medida de erro do modelo

def calculo_mape(y_pred,y_true):

    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
# Ajustar parâmetros de plotagem

plt.rcParams["text.color"] = 'Blue'
plt.rcParams["axes.titlesize"] = 20
plt.rcParams["axes.labelsize"] = 12
plt.rcParams["xtick.labelsize"] = 12
plt.rcParams["ytick.labelsize"] = 12
plt.rcParams["legend.fontsize"] = 12
plt.rcParams["figure.figsize"] = [15, 6]
plt.rcParams["lines.linewidth"] = 2.0
```

```python
import pmdarima as pm
```

### 2.1 Separando os dataframes de treino e teste

```python
# Utilizando 80% do dataset para treino e 20% para teste

train, test = serie_cluster2[:144], serie_cluster2[144:]
```

```python
len(train)
```

```python
len(test)
```

## 2.2 Treinando o modelo

---
O modelo será treinado com os seguintes parâmetros:
<ul>
    <li>m = 12, para um período de 12 meses</li>
    <li>stationary = False, uma vez que a série não é estacionária, como visto na análise exploratória</li>
    <li>seasonal = True, já que vimos que a série de arrecadação apresenta sazonalidade</li>
    <li>stepwise = False, aproveitando o reduzido tamanho da série para analisar todas as combinações de hiperparâmetros</li>
    <li>trend = t</li>
</ul>    

```python
cluster2_aarima = pm.auto_arima(train, m=12, stationary=False, seasonal=True, stepwise=False,
                           supress_warnings=True, error_action='ignore', trend='t')

cluster2_aarima.fit(train)
```

```python
cluster2_aarima.summary()
```

```python
#Faz a previsão no mesmo tamanho que a base de teste
yhat = cluster2_aarima.predict(n_periods=len(test))
yhat = pd.Series(yhat, index = test.index)
#Obtém o intervalo de confiança 
conf_int = cluster2_aarima.predict(len(test), return_conf_int=True, alpha = 0.05) 

```

```python

df_data = (pd.to_datetime(pd.DataFrame(test).index))

#Plota o resultado do modelo
fig, ax = plt.subplots()


ax.plot(test, label='Valor Esperado')
ax.plot(yhat, color='red', label='AUTO-ARIMA')
ax.fill_between(test.index, conf_int[1][:,0], conf_int[1][:,1], color='k', alpha=.08)
plt.title('Teste AUTO-ARIMA - MAPE: %.4f' % calculo_mape(test, yhat))
plt.grid(True,axis='y')
plt.xticks(ax.get_xticks(),df_data.strftime('%m/%Y'), rotation='vertical')
plt.legend(loc='upper left')
```

```python
## Formatação das tabelas

pd.set_option('max_columns', None)
pd.options.display.float_format = '{:,.2%}'.format
```

```python
variacao_percentual = pd.DataFrame(((yhat/test)-1), columns=['variacao'])

```

```python
set_alerta_negativo = pd.DataFrame()
set_normal= pd.DataFrame()
set_alerta_positivo = pd.DataFrame()
```

```python


for i in range (len(variacao_percentual)):
    if variacao_percentual.iloc[i].values[0] > 0.1:
        set_alerta_negativo = set_alerta_negativo.append(variacao_percentual[i:i+1])
    elif variacao_percentual.iloc[i].values[0] < -0.1:
        set_alerta_positivo = set_alerta_positivo.append(variacao_percentual[i:i+1])
    else:
        set_normal = set_normal.append(variacao_percentual[i:i+1])
        
```

```python
df_data = (pd.to_datetime(pd.DataFrame(test).index))

#Plota o resultado do modelo
fig, ax = plt.subplots()



ax.set_ylabel('Variação %')
plt.plot(variacao_percentual * 100, 'o', color='white')
plt.plot(set_alerta_negativo * 100, 'o', color='red',  label='Arrecadaçao < Previsão Crítico')
plt.plot(set_normal * 100, 'o', color='green',  label='Períodos Normais')
plt.plot(set_alerta_positivo * 100, 'o', color='blue',  label='Arrecadação > Previsão Crítico')
plt.title('Variação entre valores previstos e observados - AUTO-ARIMA' )
plt.grid(True,axis='y')
plt.xticks(ax.get_xticks(),df_data.strftime('%m/%Y'), rotation='vertical')
plt.legend(loc='upper right')
```

```python
#Plotagem dos resíduos e densidade dos resíduos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,4))
residuals = pd.DataFrame(cluster2_aarima.resid())
residuals.plot(ax=ax2, title='Resíduos', label=None)
residuals.plot(ax=ax1,kind='kde', title='Densidade dos resíduos')
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Plota o diagnóstico do modelo 
cluster2_aarima.plot_diagnostics()
plt.plot()
```

## 2.3 Validando a predição

```python
#Faz a previsão para a base de validacao
yhat = cluster2_aarima.predict(n_periods=len(validacao_cluster2))
yhat = pd.Series(yhat, index = validacao_cluster2.index)
#Obtém o intervalo de confiança 
conf_int = cluster2_aarima.predict(len(validacao_cluster2), return_conf_int=True, alpha = 0.05) 
```

```python
df_data = (pd.to_datetime(pd.DataFrame(validacao_cluster2).index))

#Plota o resultado do modelo
fig, ax = plt.subplots()


ax.plot(validacao_cluster2, label='Valor Esperado')
ax.plot(yhat, color='red', label='AUTO-ARIMA')
ax.fill_between(validacao_cluster2.index, conf_int[1][:,0], conf_int[1][:,1], color='k', alpha=.08)
plt.title('Teste AUTO-ARIMA - MAPE: %.4f' % calculo_mape(yhat,validacao_cluster2))
plt.grid(True,axis='y')
plt.xticks(ax.get_xticks(),df_data.strftime('%m/%Y'), rotation='horizontal')
plt.legend(loc='upper left')



```
